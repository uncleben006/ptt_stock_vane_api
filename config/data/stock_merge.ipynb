```python
import pandas as pd
import os


stock_targets = pd.read_csv("./stock_targets.csv", dtype=str)
stock_nicknames = pd.read_csv("./stock_nicknames.csv", dtype=str)
stock_infos = pd.read_csv("./stock_infos.csv", dtype=str)
stock_merge = stock_targets.merge(stock_infos, on='stock_no', how='left').merge(stock_nicknames, on='stock_no', how='left').drop(columns=['stock_name_y']).rename(columns={'stock_name_x':'stock_name'})
stock_merge.to_csv("./stock_merge.csv", index=False)
stock_merge
```
